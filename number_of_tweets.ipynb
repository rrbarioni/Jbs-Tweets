```python
import csv
import os
import numpy as np
import datetime
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
tweets_files_folder = 'tweets_1000_day/raw'
tweets_file_list = os.listdir(tweets_files_folder)
```

```python
files_dates = []
tweets_count = []
for file_name in tweets_file_list:
    with open(tweets_files_folder + '/' + file_name, encoding='utf-8') as txt_file:
        current_file_date = file_name[4:14]
        files_dates += [current_file_date]
        
        tweets = [line.strip() for line in txt_file]
        tweets_count += [len(tweets)]
```

```python
times = pd.date_range(files_dates[0], periods=365)

fig, ax = plt.subplots(figsize=(16, 7))
fig.autofmt_xdate()

plt.xlabel("Timeline")
plt.ylabel("Number of tweets")
plt.plot(times, tweets_count)
plt.show()
```
